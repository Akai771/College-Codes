```python
import numpy as np

class Perceptron:
    def __init__(self, lr=0.1, epochs=20):
        self.lr = lr
        self.epochs = epochs

    def activation(self, x):
        return np.where(x >= 0, 1, 0)          # step function

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.w = np.zeros(n_features)            # init weights = 0
        self.b = 0                               # init bias = 0

        for epoch in range(self.epochs):
            errors = 0
            for i, x_i in enumerate(X):
                net_input = np.dot(x_i, self.w) + self.b
                y_pred    = self.activation(net_input)

                error   = y[i] - y_pred              # 0=correct, ±1=wrong
                self.w += self.lr * error * x_i      # update weights
                self.b += self.lr * error             # update bias
                errors += int(error != 0)

    def predict(self, X):
        net_input = np.dot(X, self.w) + self.b
        return self.activation(net_input)
```

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,1,1,1])

p = Perceptron()
p.fit(X, y)

pred = p.predict(X)
print("Actual:   ", y)
print("Predicted:", pred)
```
