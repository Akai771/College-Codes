# Practical 12 - Aim: Training and using word embeddings

```python
import gensim
from gensim.models import Word2Vec

# import numpy as np
import pandas as pd
import re
```

```python
dataset = pd.read_csv("Corona_NLP_train.csv", encoding='latin')
print(dataset[:5])
```

```python
# clean the text by removing characters other than letters and hyperlinks

texts = []
for i in range(0, len(dataset)):
    text = re.sub('[^a-zA-Z]', ' ', dataset['OriginalTweet'][i])
    text = text.lower()
    text = text.split()
    x = len(text) if text.count('https') == 0 else text.index('https')
    text = text[: x]
    text = [t for t in text if not t == 'https']
    text = ' '.join(text)
    texts.append(text)

# check length of texts
print(len(texts))
```

```python
# print first five tweets after cleaning
# print(texts[:5])

# apply word2vec
sentences = [line.split() for line in texts]
w2v = Word2Vec(sentences, vector_size=100, epochs=10, window=5, workers=5, min_count=5, seed=2)
# print(len(w2v.wv))

# print(sentences[20:25])
words = list(w2v.wv.index_to_key)
print("printing words")
print(words[:5])

print("printing corona")
print(w2v.wv['corona'])

print("Finding similarity of vladimir putin")
print(w2v.wv.similarity('vladimir', 'putin'))

print("Finding similarity of vladimir modi")
print(w2v.wv.similarity('vladimir', 'modi'))
```

```python
print("Finding most similar of wife")
print(w2v.wv.most_similar('wife'))

print("Finding does not match")
print(w2v.wv.doesnt_match(['grocery', 'covid', 'coronavirus']))
```
