# Practical 14 - Aim: Creating a text summarization tool

```python
import sumy
import nltk
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lex_rank import LexRankSummarizer

# download required nltk library
nltk.download('punkt_tab')

text = """
Mr. Robot follows Elliot Alderson, a brilliant but socially anxious cybersecurity engineer and vigilante hacker living in New York City,
who is recruited by a mysterious anarchist known as "Mr. Robot" to join an underground hacktivist group called fsociety. Their primary
mission is to dismantle the global conglomerate E Corp by deleting all consumer debt records, an act they believe will liberate society
from corporate enslavement. However, as the plan unfolds, Elliot's tenuous grip on reality begins to fracture, revealing that he suffers 
from Dissociative Identity Disorder and that his relationship with Mr. Robot—and the true nature of their revolution—is far more complex
and personal than he initially realized. The narrative eventually shifts from a global hacktivist thriller into a deeply psychological journey
of trauma recovery, as Elliot confronts the dark secrets of his past to determine who is truly "in control" of his mind and the world's future.
"""

# Parse Text
parser = PlaintextParser.from_string(text, Tokenizer("english"))

# create summarizer
summarizer = LexRankSummarizer()

# Generate Summary
summary = summarizer(parser.document, 2)

for sentence in summary:
    print(sentence)
```
