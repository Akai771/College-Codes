# Practical 13 - Aim: Implementing a text classifier

```python
# feature extractor function
def gender_features(word):
    return {'last_letter': word[-1]}

gender_features('Shrek')
```

```python
# data set containing class labels
import nltk
from nltk.corpus import names
nltk.download('names')
import random

names = ([(name, 'male') for name in names.words('male.txt')] + 
         [(name, 'female') for name in names.words('female.txt')])

random.shuffle(names)

print(names[:10])
```

```python
# dividing dataset into train and test set
featuresets = [(gender_features(n), g) for (n, g) in names]
train_set, test_set = featuresets[500:], featuresets[:500]

classifier = nltk.NaiveBayesClassifier.train(train_set)

#testing the classifier
print(classifier.classify(gender_features('Pradumay')))

# test the accuracy on the test data
print(nltk.classify.accuracy(classifier, test_set))

# show most effective features for distinguishing the names gender
print(classifier.show_most_informative_features(5))
```
