```python
# import the necessary libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.datasets import load_iris

# create a dataframe by reading csv file
df=pd.read_csv("FeatureSelection.csv")
```

```python
#display the dataframe
df.head()
df.shape
df.describe()
```

```python
#copy the dataframe to another dataframe
df1=df.copy()
df1
```

```python
#Normalization Scaling by using min max value
for column in df1.columns:
    df1[column]=(df1[column]-df1[column].min())/(df1[column].max()-df1[column].min())
df1

df1.plot(kind='bar')
df.plot(kind='bar')
```

```python
#Standardization (z score normalization)
# new_value=(old_value-mean of column)/std of same column
df2=df.copy()
for column in df2.columns:
    df2[column]=(df2[column]-df2[column].mean())/df2[column].std()
df2
df2.plot(kind='bar')
```

```python
data=np.array([[10],[20],[30],[40]])
scaler=StandardScaler()
transformed_data=scaler.fit_transform(data)
print("Transformed Data:\n",transformed_data)
```

```python
scaler=MinMaxScaler()
scaled_data=scaler.fit_transform(data)
print("Normalized Data:\n",scaled_data)
```

```python
data=load_iris()
X=data.data#features
y=data.target#class

selector=SelectKBest(score_func=chi2,k=2)#score_fun -> function name, k-> no of features
X_new=selector.fit_transform(X,y)
print("Original Shape",X.shape)
print("Reduced shape:",X_new.shape)
```
