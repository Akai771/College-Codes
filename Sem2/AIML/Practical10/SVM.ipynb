```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
#Two classes- Red and Blue points
X=np.array([[1,2],[2,3],[2,4],[3,2],[5,6],[6,5],[6,7],[7,6]])
y=np.array([0,0,0,0,1,1,1,1])

print("Dataset ready ! shape:",X.shape)
```

```python
###Step-3 Train the SVM model
model = SVC(kernel='linear', C=1)
model.fit(X,y)

print("Model trained!")
print("Support Vectors:\n" , model.support_vectors_)
```

```python
###Step:-4 Visualise - hyperplane, M
fig, ax = plt.subplots(figsize=(8,6))


#------------plot data points-----
ax.scatter(X[y==0, 0],X[y==0, 1],color='blue', s=100, label='Class 0')
ax.scatter(X[y==1, 0],X[y==1, 1],color='red', s=100, label='Class 1')
#---Draw hyperplane and margins---
w = model.coef_[0]   #weight vector
b = model.intercept_[0]  #bias

x_vals = np.linspace(0, 8, 100)

#Hyperplane:      w[0]*x + w[1]*y +b => y = -(w[0]*X =b) / w[1]
hyperplane = -(w[0] * x_vals + b) / w[1]
margin_top = -(w[0] * x_vals + b - 1) / w[1]  # +1 margin
margin_bot = -(w[0] * x_vals + b + 1) / w[1]  # -1 margin

ax.plot(x_vals, hyperplane, 'k-', lw=2, label = 'Hyperplane') 
ax.plot(x_vals, margin_top, 'r--', lw=1.5, label = 'Margin (+1)')
ax.plot(x_vals, margin_bot, 'b--', lw=1.5, label = 'Margin(-1)')

#----Highlight Support Vectors----
ax.scatter(model.support_vectors_[:, 0],
           model.support_vectors_[:, 1],
           s=300, facecolors= 'none', edgecolors= 'green',
           linewidths = 2.5, label = 'Support Vectors')

#-----labels-----
ax.set_xlim(0,8)
ax.set_ylim(0,9)
ax.set_xlabel("feature 1")
ax.set_ylabel("feature 2")
ax.set_xlabel("SVM : hyperplane, margin & support Vectors")
ax.legend()
ax.grid(True,alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Margin width = {2 / np.linalg.norm(w):.3f}")
```
