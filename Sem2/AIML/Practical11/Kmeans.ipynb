```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv("income - income.csv")
df
```

```python
plt.scatter(df['age'],df['income'])
```

```python
## Apply K-means
kmean=KMeans(n_clusters=3)
kmean

y_predict=kmean.fit_predict(df[['age','income']])
y_predict

df['cluster']=y_predict
df

kmean.predict([[34,50000]])

#divide the dataframe according to the cluster
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
```

```python
plt.scatter(df1.age,df1.income,color='blue')
plt.scatter(df2.age,df2.income,color='green')
plt.scatter(df3.age,df3.income,color='red')
plt.xlabel('age')
plt.xlabel('income')
```

```python
scaler=MinMaxScaler() #scaler is MinMaxScaler model
scaler.fit(df[['age']]) #fit() is used to train the model

df['AGE']=scaler.transform(df[['age']])
df

scaler.fit(df[['income']])

df['Income']=scaler.transform(df[['income']])
df
```

```python
km=KMeans(n_clusters=3) # generate kmeans model
y_predict=km.fit_predict(df[['AGE','Income']])
y_predict
```

```python
df['cluster']=y_predict
df.head()

df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]

plt.scatter(df1.AGE,df1.Income,color='blue')
plt.scatter(df2.AGE,df2.Income,color='green')
plt.scatter(df3.AGE,df3.Income,color='red')
plt.xlabel('age')
plt.xlabel('income')
```

```python
#display cluster means
km.cluster_centers_
```

```python
plt.scatter(df1.AGE,df1.Income,color='blue')
plt.scatter(df2.AGE,df2.Income,color='green')
plt.scatter(df3.AGE,df3.Income,color='red')
plt.scatter(km.cluster_centers_[:,0],
            km.cluster_centers_[:,1],marker='d')
```

## Elbow  Method

```python
k_range=range(1,10)
sse=[]
for k in k_range:
    km=KMeans(n_clusters=k)
    km.fit(df[['AGE','Income']])
    sse.append(km.inertia_)
sse #noise created
```

```python
plt.xlabel('k')
plt.ylabel("Sum of square error")
plt.plot(k_range,sse)
```
