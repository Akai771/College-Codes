```python
import numpy as np

class Adaline:
    def __init__(self, lr=0.1, epochs=100):
        self.lr = lr
        self.epochs = epochs

    def net_input(self, X):
        return np.dot(X, self.w) + self.b      # linear combination

    def fit(self, X, y):
        n_features = X.shape[1]
        self.w = np.zeros(n_features)           # init weights = 0
        self.b = 0                               # init bias = 0

        for epoch in range(self.epochs):
            total_err = 0
            for i in range(len(X)):
                output = self.net_input(X[i])   # linear output (NOT step)
                error  = y[i] - output           # continuous error
                self.w += self.lr * error * X[i]
                self.b += self.lr * error
                total_err += error ** 2          # sum of squared errors

            # if (epoch + 1) % 20 == 0:
            #     print(f"Epoch {epoch+1:3d} | SSE={total_err:.4f} | w={self.w} | b={self.b:.3f}")

    def predict(self, X):
        return np.where(self.net_input(X) >= 0.5, 1, 0)   # threshold at test time
```

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])    # AND gate labels

model = Adaline(lr=0.1, epochs=100)
model.fit(X, y)

pred = model.predict(X)
print("\nInput  | Actual | Predicted")
for i in range(len(X)):
    print(f"{X[i]}  |   {y[i]}    |   {pred[i]}")
```
