```python
# not a number- nan()
df1 = pd.DataFrame(
    data=[[1, np.nan, 2], [4, 5, 6]], index=["r1", "r2"], columns=["c1", "c2", "c3"]
)
df1
```

```python
df1.isnull()  # returns true for nan cell
```

```python
# replace
df1.replace(to_replace="Nan", value=3)
```

```python
df1.replace(to_replace=np.nan, value=3)
```

```python
df = pd.DataFrame(
    {
        "Name": ["Alice", "Bob", "Charlie", "Diana"],
        "Age": [22, np.nan, 21, np.nan],
        "Marks": [85, 92, np.nan, 78],
    }
)
df
```

```python
# fill all nan with number
print(df.fillna(0))
```

```python
# fill nan with string
print(df.fillna("Unknown"))
```

```python
# count number of nan cells
df.isnull().sum()  # counts columns wise
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 4, 7]
```

```python
plt.plot(x, y1, color="orange", label="Line1", marker="o")
```

```python
plt.plot(x, y1, color="Blue", label="Line1", marker="o")
plt.plot(x, y2, color="orange", label="Line2", marker="s")
plt.title("Multiple Lines")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# pie chart
sizes = [35, 25, 15, 5, 10]
labels = ["Python", "Java", "C++", "JS", "Other"]
plt.pie(
    sizes,
    labels=labels,
    autopct="%1.1f%%",
    explode=[0.05, 0, 0, 0, 0],
    colors=["blue", "green", "yellow", "orange", "red"],
    shadow=True,
    startangle=90,
    counterclock=False,
    wedgeprops={"edgecolor": "white", "linewidth": 2},
)
plt.show()
```
