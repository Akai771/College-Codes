```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import plot_tree
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
iris = load_iris()
X, y = iris.data, iris.target
class_names = iris.target_names

print("Shape:", X.shape)               # (150, 4)
print("Classes:", class_names)          # setosa, versicolor, virginica
print("Features:", iris.feature_names)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
rf = RandomForestClassifier(n_estimators=20, random_state=42)  # 20 trees
rf.fit(X_train, y_train)
```

```python
fig, axs = plt.subplots(1, len(rf.estimators_), figsize=(20, 5))
for i, tree in enumerate(rf.estimators_):
    plot_tree(tree,
              feature_names=iris.feature_names,
              class_names=class_names,
              filled=True,
              ax=axs[i])
    axs[i].set_title(f"Tree {i+1}")
plt.tight_layout()
plt.show()
```

```python
# Single sample — shows majority voting across all trees
s = X_test[10].reshape(1, -1)
tree_votes = [int(tree.predict(s)[0]) for tree in rf.estimators_]

print("Tree votes:", tree_votes)
print("Predicted: ", class_names[rf.predict(s)[0]])
print("Actual:    ", class_names[y_test[10]])
```

```python
y_predicted = rf.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_predicted):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_predicted, target_names=class_names))
```

```python
cm = confusion_matrix(y_test, y_predicted)
print(cm)
```

```python
# Plot the confusion matrix
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```
